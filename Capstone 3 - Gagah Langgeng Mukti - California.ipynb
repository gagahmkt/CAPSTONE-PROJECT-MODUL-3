Distribusi data menggunakan histogram yang menunjukkan data tidak normal sehingga harus dilakukan handling outliers.

```python
# Mencari outliers

columns = ['longitude', 'latitude','housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

for i in columns:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1

    # Mengidentifikasi batas atas dan batas bawah
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]

    print(f'Feature: {i}')
    print(f'IQR: {IQR}')
    print(f'Lower bound: {lower_bound}')
    print(f'upper bound: {upper_bound}')
    print(f'Total outliers: {outliers[i].count()}\n----------------------\n')
```

### CLEANED DATAFRAME

Dataframe yang sudah difilter sehingga outliers tidak dimasukkan ke dataframe

```python
columns = df[['longitude', 'latitude','housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']]

for i in columns:
    df_cleaned = df[(df[i] >= lower_bound) & (df[i] <= upper_bound)]
```

# **3. DATA PROCESSING**

## FEATURE ENGINEERING

Feature engineering adalah proses penerapan pengetahuan domain dan pemahaman data untuk membuat atau mengekstraksi fitur-fitur baru yang lebih informatif dari data mentah. Tujuan utama dari feature engineering adalah untuk meningkatkan kinerja model machine learning dengan menyediakan informasi yang lebih relevan dan bermakna kepada model.

> Encoding

Encoding digunakan untuk mengonversi data kategori atau teks menjadi format yang dapat diproses oleh algoritma machine learning yang berbasis matematika. Tujuan utamanya adalah untuk mengubah data kategorikal menjadi representasi numerik yang dapat digunakan oleh model untuk pembelajaran. Teknik encoding pada data ini adalah menggunakan One Hot Encoding. 

>Scaling

Scaling dilakukan untuk memastikan bahwa feature dalam dataset memiliki skala yang sama. Bertujuan untuk mengoptimalkan kinerja algoritma machine learning yang sensitif terhadap perbedaan skala antar feature. Scaling yang digunakan pada machine learning ini adalah Robust Scaler, karena ideal digunakan pada data yang berdistribusi tidak normal dan memiliki outliers

>Metrics

Evaluation Metrics yang digunakan pada kasus ini adalah Mean Absolute Percentage Error (MAPE). Mean Absolute Percentage Error (MAPE) adalah metrik evaluasi yang sering digunakan dalam kasus dimana ingin mengevaluasi seberapa baik model regresi memprediksi nilai dalam skala persentase.

MAPE digunakan sebagai metrics evaluasi pada machine learning karena:
- Interpretasi yang Mudah: MAPE memberikan hasil dalam bentuk persentase, yang mudah diinterpretasikan. Berguna untuk mengetahui seberapa akurat model dalam memprediksi nilai dalam skala persentase dari nilai sebenarnya.

- Menggambarkan Tingkat Kesalahan Relatif: MAPE mengukur kesalahan relatif dari prediksi, yang memberikan gambaran tentang seberapa besar kesalahan dalam prediksi dibandingkan dengan nilai sebenarnya.

- Sensitif terhadap Kesalahan yang Besar: MAPE sensitif terhadap kesalahan yang besar, karena kesalahan dinyatakan dalam bentuk persentase. Berguna dalam menilai kualitas prediksi model dalam kasus dimana kesalahan yang besar dianggap lebih kritis.

### DEFINE FEATURE AND TARGET

Melakukan pemilihan feature (X) dan target (y) yang akan digunakan pada Machine Learning

```python
# Mendefinisikan X dan Y

X = df_cleaned.drop(columns='median_house_value')
y = df_cleaned['median_house_value']
```

```python
df_cleaned.columns
```

Dari keseluruhan kolom pada DataFrame, kolom yang digunakan sebagai feature adalah 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', dan 'ocean_proximity', serta'median_house_value' sebagai target yang akan digunakan untuk prediksi harga rumah di kemudian hari.

```python
# Splitting data

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size= 0.2,
    random_state= 2019
)
```

```python
# Membuat pipeline untuk preprocessing numerik
numeric_features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']
numeric_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

# Membuat pipeline untuk preprocessing kategorikal
categorical_features = ['ocean_proximity']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop = 'first'))
])

# Menggabungkan kedua pipeline untuk seluruh fitur
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ], remainder='passthrough')
```

```python
# Model yang digunakan

lin = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor()
svr = SVR()
rf = RandomForestRegressor()
ab = AdaBoostRegressor()
lasso = Lasso()
ridge = Ridge()
cb = CatBoostRegressor(verbose= False)
xg = XGBRegressor()
```

# **4. MODELING**

## MODEL BENCHMARKING

```python
# Define Model
models = [lin, knn, dt, svr, rf, ab, lasso, ridge, cb, xg]
mape = []
mean_mape = []
std_mape = []

# Perform cross-validation tiap model menggunakan MAPE
for model in models:
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    # model_cv = cross_val_score(pipeline, X, y, cv=5, scoring= make_scorer(mean_absolute_percentage_error, greater_is_better=False))  # Using 5-fold cross-validation and MAPE as scoring metric)  # Using 5-fold cross-validation and R-squared as scoring metric
    model_cv = cross_val_score(pipeline, X_train, y_train, cv = 5, scoring = 'neg_mean_absolute_percentage_error')
    model_cv = -model_cv
    mape.append(model_cv)
    mean_mape.append(model_cv.mean())
    std_mape.append(model_cv.std())
```

```python
# Define Model
models = [lin, knn, dt, svr, rf, ab, lasso, ridge, cb, xg]
mae = []
mean_mae = []
std_mae = []

# Perform cross-validation tiap model menggunakan MAE
for model in models:
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    # model_cv = cross_val_score(pipeline, X, y, cv=5, scoring= make_scorer(mean_absolute_percentage_error, greater_is_better=False))  # Using 5-fold cross-validation and MAPE as scoring metric)  # Using 5-fold cross-validation and R-squared as scoring metric
    model_cv = cross_val_score(pipeline, X, y, cv = 5, scoring = 'neg_mean_absolute_error')
    model_cv = -model_cv
    mae.append(model_cv)
    mean_mae.append(model_cv.mean())
    std_mae.append(model_cv.std())
```

```python
# Membuat dataframe untuk hasil cross validation

df_cv = pd.DataFrame({
    'Model' : ['Linear Regression', 'KNN', 'Decision Tree', 'SVR', 'Random Forest', 'AdaBoost', 'Lasso', 'Ridge', 'CatBoost', 'XGBoost'],
    'Avg MAPE' : mean_mape,
    'Std MAPE' : std_mape,
    'AVG MAE' : mean_mae,
    'Std MAE' : std_mae
})

df_cv.sort_values('Avg MAPE')
```

Dipilih 3 model terbaik setelah dilakukan model benchmarking terhadap model-model di atas berdasarkan Avg MAPE terkecil, antara lain CatBoost Regressor, XGBoost Regressor, dan Random Forest Regressor. Selanjutnya dilakukan hyperparameter tuning terhadap 3 model terpilih.

## HYPERPARAMETER TUNING

Hyperparameter tuning dilakukan pada 3 model tereatas dari hasil cross validation model, yaitu CatBoost Regressor, XGBoost Regressor, dan Random Forest Regressor.

Hyperparameter tuning dilakukan sebagai usaha untuk memperbaiki hasil dari model-model yang digunakan untuk menemukan satu model terbaik.

> CatBoost Regressor

```python
# Parameter untuk CatBoost Regressor

cb_param_dist = {
    'model__learning_rate': uniform(0.01, 0.1),
    'model__depth': randint(1, 10)
}

# Define pipeline for CatBoost Regressor

cb_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', cb)
])

# Perform RandomizedSearchCV untuk CatBoost Regressor

cb_random = RandomizedSearchCV(estimator=cb_pipe, param_distributions=cb_param_dist, n_iter=30, scoring='neg_mean_absolute_percentage_error', cv=5, random_state=2019)

# Fit the model

cb_random.fit(X_train, y_train)

# Menampilkan hyperparameter dan MAPE terbaik

print("Hyperparameters terbaik untuk CatBoost Regressor:", cb_random.best_params_)
print("Mean Absolute Percentage Error (MAPE) CatBoost Regressor:", -cb_random.best_score_)
```

> XGBoost Regressor

```python
# Parameter untuk XGBoost Regressor

xg_param_dist = {
    'model__learning_rate': uniform(0.01, 0.1),
    'model__max_depth': randint(1, 10),
    'model__min_child_weight': randint(1, 10),
    'model__subsample': uniform(0.5, 0.5),
    'model__colsample_bytree': uniform(0.5, 0.5)
}

# Pipeline untuk XGBoost Regressor

xg_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', xg)
])

# Randomized Search

xg_random = RandomizedSearchCV(xg_pipe, xg_param_dist, n_iter=30, cv=5, scoring='neg_mean_absolute_percentage_error', random_state=2019)

# Fit randomized search

xg_random.fit(X_train, y_train)

# Menampilkan hyperparameter dan MAPE terbaik

print("Hyperparameters terbaik untuk XGBoost Regressor:", xg_random.best_params_)
print("Mean Absolute Percentage Error (MAPE) XGBoost Regressor:", -xg_random.best_score_)
```

> Random Forest Regressor

```python
# Parameter untuk Random Forest Regressor

rf_param_dist = {
    'model__n_estimators': randint(100, 500),
    'model__criterion' : ['squared_error'],
    'model__max_depth': randint(1, 10),
    'model__min_samples_split': randint(1, 10),
    'model__min_samples_leaf': randint(1, 10)
}

# Pipeline untuk Random Forest Regressor

rf_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', rf)
])

# Randomized Search

rf_random = RandomizedSearchCV(estimator=rf_pipe, param_distributions=rf_param_dist, n_iter=30, cv=5, scoring='neg_mean_absolute_percentage_error', random_state=2019)

# Fit randomized search

rf_random.fit(X_train, y_train)

# Menampilkan hyperparameter dan MAPE terbaik

print("Hyperparameters terbaik untuk Random Forest Regressor:", rf_random.best_params_)
print("Mean Absolute Percentage Error (MAPE) Random Forest Regressor:", -rf_random.best_score_)
```

```python
# Fit model pipeline dengan training data

cb_best = CatBoostRegressor(learning_rate= 0.09892146954077474, depth= 7, verbose= False)

cb_best_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', cb_best)
])

cb_best_pipe.fit(X_train, y_train)

# Prediksi pada data test

y_pred = cb_best_pipe.predict(X_test)

# MAPE pada data test

mape_test = mean_absolute_percentage_error(y_test, y_pred)

print('MAPE pada data test:', mape_test)
```

Hasil MAPE yang diterapkan pada data test sebesar 0.1691, menunjukkan bahwa hasil model CatBoost Regresor pada data train dan data tes tidak jauh berbeda.
